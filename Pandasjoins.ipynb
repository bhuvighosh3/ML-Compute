```python
player=pd.read_csv('/content/player.csv')
```

```python
#player.columns
```

```python
player_attributes=pd.read_csv('/content/player_attributes.csv')
```

```python
#player_attributes.columns
```

```python
player_new=pd.merge(left=player,right=player_attributes,on='id',how='right',suffixes=('', '_drop')).filter(regex='^(?!.*_drop)')
```

```python
#player_new.columns
```

```python
team=pd.read_csv('/content/teams.csv')
```

```python
#team.columns
```

```python
team_attributes=pd.read_csv('/content/team_attributes.csv')
```

```python
#team_attributes.columns
```

```python
team_new=pd.merge(left=team,right=team_attributes,on='team_fifa_api_id',how='left',suffixes=('', '_drop')).filter(regex='^(?!.*_drop)')
```

```python
#team_new.columns
```

```python
match=pd.read_csv('/content/matches.csv')
```

```python
#match.columns
```

```python
league=pd.read_csv('/content/leagues.csv')
```

```python
#league.columns
```

```python
league.rename(columns={'name':'Leaguename'},inplace=True)
```

```python
match_new=pd.merge(left=match,right=league,on='id',how='right',suffixes=('', '_drop')).filter(regex='^(?!.*_drop)')
```

```python
#match_new.columns
```

```python
country=pd.read_csv('/content/countries.csv')
```

```python
#country.columns
```

```python
teams=pd.merge(left=country,right=team_new,on='id',how='left',suffixes=('', '_drop')).filter(regex='^(?!.*_drop)')
```

```python
#teams.columns
```

```python
teamsplayers=pd.merge(left=teams,right=player_new,on='id',how='left',suffixes=('', '_drop')).filter(regex='^(?!.*_drop)')
```

```python
final=pd.merge(left=match_new,right=teamsplayers,on='id',how='left',suffixes=('', '_drop')).filter(regex='^(?!.*_drop)')
```

```python
df_final=final[['name','team_long_name','season','Leaguename','home_team_goal','away_team_goal']]
```
